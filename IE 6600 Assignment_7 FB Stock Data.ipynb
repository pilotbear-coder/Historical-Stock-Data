#### 2. Create a subplot layout consisting of four subplots with 2 rows and 2 columns to visualize Facebook's 2018 stock data: open, high, low, and close prices. Set the figure size as 14 by 6. Set the main title as "Facebook 2018 Stock Data," the x-label as "Date," and the y-label as "Price ($)" for the entire visualization. (15 Points)

```python
# GRADED Task 2: Please add your code below in the `your code here` sections

# Create a subplot layout with four subplots (open, high, low, and close prices) using a 2x2 grid.
# Please store your plot in task_2_plot =  and do not remove this value.

task_2_plot = plt.subplots(2, 2, figsize=(14, 6))
fig, axes = task_2_plot 
axes[0, 0].plot(fb.index, fb['open'])
axes[0, 0].set_title('Open Price')


axes[0, 1].plot(fb.index, fb['high'])
axes[0, 1].set_title('High Price')


axes[1, 0].plot(fb.index, fb['low'])
axes[1, 0].set_title('Low Price')


axes[1, 1].plot(fb.index, fb['close'])
axes[1, 1].set_title('Close Price')

# Set the main title for the entire subplot layout.
fig.suptitle('Facebook 2018 Stock Data')

# Set the label for the x-axis (shared by all subplots).
fig.text(0.5, 0.04, 'Date', ha='center', fontsize=12)

# Set the label for the y-axis (shared by all subplots).
fig.text(0.04, 0.5, 'Price ($)', va='center', rotation='vertical', fontsize=12)

# Display the subplot layout.
plt.tight_layout()
plt.subplots_adjust(bottom=0.1, left=0.1)

plt.show()

```

#### 3. Create a line plot to visualize the 'open,' 'high,' 'low,' and 'close' prices of Facebook stock data. Set the xlabel to "Date" and the ylabel to "Price ($)". Set the title as "Facebook Stock Prices in 2018." Configure the figure size as 16 by 6. Add a legend with the following parameters: location as "upper left," title for the legend as "Stock," and arrange the legend items in a single row. Additionally, ensure that all text elements in the plot, including labels, title, and legend, have a font size of 14 for improved readability. (20 Points)

```python
# GRADED Task 3: Please add your code below in the `your code here` sections

# Create a line plot for the 'open,' 'high,' 'low,' and 'close' prices of Facebook stock with the figure size set to 16 by 6.
# Please store your plot in task_3_plot and do not remove this value.
task_3_plot = plt.figure(figsize=(16, 6))
plt.plot(fb.index, fb['open'], label='Open')
plt.plot(fb.index, fb['high'], label='High') 
plt.plot(fb.index, fb['low'], label='Low')
plt.plot(fb.index, fb['close'], label='Close')

# Set the title
plt.title('Facebook Stock Prices 2018')

# Set the label for the x-axis.
plt.xlabel('Date')

# Set the label for the y-axis.
plt.ylabel('Price ($)')

# Add a legend to the plot
plt.legend(loc='upper left', title='Stock', ncol=4, fontsize=14, title_fontsize=14)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.show()

```

#### 6. Use the axvspan() function to shade a rectangle from '2018-07-25' to '2018-07-31' and label it as "July Decline," marking the significant decline in Facebook's closing price on a line plot. Set the x-label to "Date," the y-label to "Price ($)," and the main title to "Facebook Closing Price." Set the figure size to 10 by 6. (20 Points)

```python
# GRADED Task 6: Please add your code below in the `your code here` sections
# Please store your plot in task_6_plot and do not remove this value.

# Create a line plot for Facebook's closing stock price with specified figure size.
task_6_plot = plt.figure(figsize=(10, 6))

# Plot the closing price as a line
plt.plot(fb.index, fb['close'], linewidth=1.5, color='blue')

# Shade the rectangle from '2018-07-25' to '2018-07-31' and label it
from datetime import datetime
start_date = datetime.strptime('2018-07-25', '%Y-%m-%d')
end_date = datetime.strptime('2018-07-31', '%Y-%m-%d')
plt.axvspan(start_date, end_date, alpha=0.3, color='red', label='July Decline')

# Set labels and title
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.title('Facebook Closing Price')

# Add legend to show the July Decline label
plt.legend()

# Improve layout
plt.tight_layout()

# Display the plot
plt.show()

```
